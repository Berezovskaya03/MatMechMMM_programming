сравнение алгоритмов SVM и kNN

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("iris.csv")
df.head()
```

```python
df = df.replace({"variety":  {"Setosa":1,"Versicolor":2, "Virginica":3}})
df.head()
```

```python
x = df.iloc[:, :-1].values 
y = df.iloc[:, 4].values
```

knn

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20)
```

```python
from sklearn.neighbors import KNeighborsClassifier 
method1 = KNeighborsClassifier()
method1.fit(x_train, y_train)
```

```python
y_pred1 = method1.predict(x_test)
```

svm

```python
from sklearn.svm import SVC
method2 = SVC(kernel = 'linear', random_state = 0)
method2.fit(x_train, y_train)
y_pred2 = method2.predict(x_test)
```

сравнение

```python
comparison_df = pd.DataFrame(data={'actual':y_test, 'model_1':y_pred1, 'model_2':y_pred2})
comparison_df
```

```python
success_1, success_2 = 0, 0
for index, row in comparison_df.iterrows():
  if row['actual'] == row['model_1']:
    success_1 += 1
  if row['actual'] == row['model_2']:
    success_2 += 1
print(f'Попали {success_1} из {len(y_test)} раз первой моделью')
print(f'Попали {success_2} из {len(y_test)} раз второй моделью')
```
